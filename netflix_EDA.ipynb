**2.** *what are different types of Genres*

*Lets first divide our Dataframe into two which are  df_movies and df_series*

```python
df_movies = df[df['Series or Movie'] == 'Movie']

df_movies.head(1)
```

```python
df_series = df[df['Series or Movie'] == 'Series']

df_series.head(1)
```

*When we want to watch some movie or Series we can filter it using Genres,so lets find out differnt and most popular Genres*

*Genre is the term for any category of creative work, which includes literature and other forms of art or entertainment (e.g. music)—whether written or spoken, audio or visual—based on some set of stylistic criteria. Genres are formed by conventions that change over time as new genres are invented and the use of old ones are discontinued. Often, works fit into multiple genres by way of borrowing and recombining these conventions.*

![Alt text](Genres.jpg)

*2.1) Different types of generes in Series category*

```python
df_series['Genre'].unique()
```

*Top Generes in Series categrory*

```python
df_series_gen = df_series.dropna(subset=['Genre']) 
```

```python
series_gen_list = df_series_gen.Genre.str.split(',') 
s_gen_list = {} 

#using Loop to create and seperate genres 

for genres in series_gen_list: 
    for genre in genres: 
        if (genre in s_gen_list): 
            s_gen_list[genre]+=1 
        else:  
            s_gen_list[genre]=1 

s_gen_df = pd.DataFrame(s_gen_list.values(),index = s_gen_list.keys(),columns = {'Counts of Genres in Tv Series'}) # type: ignore

s_gen_df.sort_values(by = 'Counts of Genres in Tv Series',ascending = False,inplace = True) #Sort the dataframe in ascending order
top_10_s_gen = s_gen_df[0:10] 
```

*Plotting Bar plot*

```python
plt.figure(figsize = (10,8))
plt.title("Top Genres in Series category")

sns.barplot( x = top_10_s_gen.index,y = top_10_s_gen['Counts of Genres in Tv Series'],)
```

*We can see that "Drama" category is most watched followed with "Animation" , "Comedy"*

*2.2) Top Genres in Movie Category* 

```python
df_movie_gen = df_movies.dropna(subset=['Genre']) 
```

```python
movie_gen_list = df_movie_gen.Genre.str.split(', ') 

m_gen_list = {} 

#using loop in seperate genre and collect them  

for genres in movie_gen_list: 
    for genre in genres: 
        if (genre in m_gen_list): 
            m_gen_list[genre]+=1 
        else:  
            m_gen_list[genre]=1 

m_gen_df = pd.DataFrame(m_gen_list.values(),index = m_gen_list.keys(),columns = {'Counts of Genres in Movies'})  # type: ignore
m_gen_df.sort_values(by = 'Counts of Genres in Movies',ascending = False,inplace = True) 

top_10_m_gen = m_gen_df[0:10] 
```

*ploting Bar Plot*

```python
plt.figure(figsize=(10,8))
plt.title("Top Genres in Movie category")

sns.barplot( x = top_10_m_gen.index,y = top_10_m_gen['Counts of Genres in Movies'])

```

*We can see that here "Drama" is most popular Genre followed by "Comedy" , "Thriller"*

**3.)What are the most popular language**

3.1) Movies

```python
df_movies_lang  = df.dropna(subset= ['Languages'])
```

```python
movie_lang_list = df_movies_lang.Languages.str.split(',')

m_lang_list = {}

for languages in movie_lang_list:
    for language in languages:
          if (language in m_lang_list): 
            m_lang_list[language]+=1
          
          else:  
            m_lang_list[language]=1 


m_lang_df = pd.DataFrame(m_lang_list.values(), index = m_lang_list.keys(),columns= {'Counts of Languages in Movies'}) #type: ignore
m_lang_df.sort_values(by = 'Counts of Languages in Movies',ascending=False,inplace=True) 


top_10_lang_movies = m_lang_df[0:10]


```

*Ploting Bar Plot*

```python
plt.figure(figsize=(10,8))
plt.title("Top languages in movies")

sns.barplot( x = top_10_lang_movies.index , y = top_10_lang_movies['Counts of Languages in Movies'])
```

As we can see that English language is most popular follwed by Japanese and Korean

*3.2)Series*

```python
df_series_lang = df.dropna(subset=['Languages'])
```

```python
series_lang_list = df_series_lang.Languages.str.split(',')


s_lang_list = {}

for languages in series_lang_list:
    for language in languages:
        if (language in s_lang_list):
            s_lang_list[language] +=1
        else:
            s_lang_list[language] = 1

s_lang_df = pd.DataFrame(s_lang_list.values(),index = s_lang_list.keys(),columns = {'Counts of Languages in Series'}) #type: ignore
s_lang_df.sort_values(by= 'Counts of Languages in Series',ascending= False,inplace = True)

top_10_lang_series = s_lang_df[0:10]
```

*Ploting Bar Plot*

```python
plt.figure(figsize=(10,8))
plt.title("Top languages in series ")

sns.barplot(x = top_10_lang_series.index , y = top_10_lang_series['Counts of Languages in Series'])
```

*We can see that English is  most viewed followed by Japanese and then Korean*
